```python
#Parsing HTML Tables in Python with BeautifulSoup and pandas
import pandas as pd
#Using Requests to Access a Web Content
import requests
from requests import get
#Parsing a Table in BeautifulSoup
from bs4 import BeautifulSoup
```

```python
url = 'https://hpd.nlm.nih.gov/cgi-bin/household/prodtree?prodcat=Personal+care&purpose=toothpaste&type=Oral+Hygiene'
# Access the HTML with the text property
response = get(url)
print(response.text[:500])

soup = BeautifulSoup(response.text, 'html.parser')
type(soup)
```

```python
tp_dict = {'name': [], 'link': []}
for i in soup.findAll('td', class_='bodytext')[6].findAll('a'):
    tp_dict['name'].append(i.text)
    tp_dict['link'].append(i.attrs['href'])
```

```python
df_parent = pd.DataFrame(tp_dict, columns=['name', 'link'])
df_parent
```
