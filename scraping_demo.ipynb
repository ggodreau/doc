```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
for i in soup.findAll('div', attrs={'class': 'articleset listings colset_4'})[0].findAll('article', attrs={'class': 'listing'}):
    print(i.findAll('ul', attrs={'class': 'dataset'}))
```

```python
df_dict = {}
```

```python
dataHeader = ('address', 'location', 'mlsid', 'status')

for dataName in dataHeader:
    df_dict[dataName] = []
    for i in soup.findAll('div', {'class': 'articleset listings colset_4'})[0]\
        .findAll('li', {'class': 'data-' + dataName}):
            df_dict[dataName].append(i.text.strip())
```

```python
pd.DataFrame.from_dict(df_dict).head(10)
```
